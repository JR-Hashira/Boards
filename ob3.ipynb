1. Visualize the survival rate of passengers based on their ticket class (1st, 2nd, or 3rd class).
Were the passengers in higher classes more likely to survive?

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Example DataFrame
data = {
    'class': ['1st', '1st', '2nd', '2nd', '3rd', '3rd', '1st', '3rd'],
    'survived': [1, 1, 1, 0, 0, 0, 1, 0]  # 1 for survived, 0 for did not survive
}

df = pd.DataFrame(data)

survival_rates = df.groupby('class')['survived'].mean() * 100  # Convert to percentage

# Optional: Set the style
sns.set(style="whitegrid")

# Create a bar plot
plt.figure(figsize=(8, 5))
sns.barplot(x=survival_rates.index, y=survival_rates.values, palette='pastel')

# Add titles and labels
plt.title('Survival Rates by Ticket Class')
plt.xlabel('Ticket Class')
plt.ylabel('Survival Rate (%)')
plt.ylim(0, 100)  # Set y-axis limits from 0 to 100

# Show the plot
plt.show()
```

3. Generate a bar chart to compare the survival rates of male and female passengers. Did gender play a role in survival?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Loads the Titanic dataset
titanic = sns.load_dataset("titanic")

# Check for missing values in the 'sex' and 'survived' columns
print(titanic['sex'].isnull().sum())
print(titanic['survived'].isnull().sum())

# Calculates and groups the survival rate by gender
survival_rates = titanic.groupby('sex')['survived'].mean() * 100

# Optional: Set the style
sns.set(style="whitegrid")

# Creates a bar plot(Source: seaborn)
plt.figure(figsize=(8, 5))
sns.barplot(x=survival_rates.index, y=survival_rates.values, palette='pastel')

# Add titles and labels
plt.title('Survival Rates by Sex')
plt.xlabel('Gender')
plt.ylabel('Survival Rate (%)')
plt.ylim(0, 100)  # Set y-axis limits from 0 to 100

# Show the plot
plt.show()


```

Explanation here:

There were a higher survival rate of women than men. If they facts are true, while evatuated the passengers to safety the women and children had to go first to the life rafts.

4. Visualize the distribution of ticket fares for each passenger class using box plots. How did ticket fares vary across different classes?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Loads the Titanic dataset
titanic = sns.load_dataset("titanic")

# Check for missing values in the 'fare' column
print(titanic['fare'].isnull().sum())

# Style
sns.set(style="whitegrid")

# Creates a plot(Source: seaborn)
plt.figure(figsize=(10, 6))
sns.boxplot(x='class', y='fare', data=titanic, palette='pastel')

# labels for axis
plt.title('Distribution of Ticket Fares base on Class')
plt.xlabel('Class Status')
plt.ylabel('Ticket Fare Sales')

# Shows the plot
plt.show()
```

5. Create a grouped bar chart to visualize the survival rates of passengers across different age groups and genders. Were certain age groups or genders more likely to survive?

```python
 import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Titanic dataset
titanic = sns.load_dataset("titanic")

# System looks missing values in the 'age' and 'survived' columns
print(titanic['age'].isnull().sum())
print(titanic['survived'].isnull().sum())

# grouping by age
bins = [0, 12, 18, 30, 50, 80]  # Define age bins
labels = ['0-12', '13-18', '19-30', '31-50', '51-80']  # Age group labels
titanic['age_group'] = pd.cut(titanic['age'], bins=bins, labels=labels, right=False)

# Calculate survival rates by age group and gender
survival_rates = titanic.groupby(['age_group', 'sex'])['survived'].mean().reset_index()
survival_rates['survival_rate'] = survival_rates['survived'] * 100

# Style
sns.set(style="whitegrid")

# Create a grouped bar plot
plt.figure(figsize=(12, 6))
sns.barplot(data=survival_rates, x='age_group', y='survival_rate', hue='sex', palette='pastel')

# Labels
plt.title('Survival Rates by Age Group and Gender on the Titanic')
plt.xlabel('Age Group')
plt.ylabel('Survival Rate (%)')
# y-axis limits are set to go from 0 to 100
plt.ylim(0, 100)

# Show the plot
plt.legend(title='Sex')
plt.show()
```

Explanation here:

In all age groups the women outsivired the men. Whille the preteens had a closet survial rate. The ages from 13 - 80 the mens survival rate is around 20% while women's is the 70s.
