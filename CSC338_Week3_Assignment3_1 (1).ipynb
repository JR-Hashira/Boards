1. A companyâ€™s HR department wants to estimate the probability of employees completing a training course. The probability of an employee completing the course is 0.8, and there are 10 employees.

```python
# The code below calculates uses binomial distribution to find the probabilities for different numbers of employees completing the course

from scipy.stats import binom

n = 10
p = 0.8

x = range(0,11,1)
print(x)

probabilities = binom.pmf(x, n, p)

# To unpack the list generated from the range and display index and values, we use ennumerate
for i, prob in enumerate(probabilities):
    print(f"(Probability of {i} employees completing the course): {prob:.3f}")
```

What does the output tell you about the likelihood of all employees completing the course versus only a few completing it?

This is a relatively low probability, indicating that it's not very likely for all employees to complete the course. However, the highest likely canidates to finish the training program would be between 7 - 9.

3. The number of customers arriving at a cafe follows a Poisson distribution with an average arrival rate of 5 customers per hour.

```python
# Finish the code to calculate the probability of exactly 3 customers arriving in a given hour.

from scipy.stats import poisson

rate = 5
k = 3

# Complete this code to calculate the probability
probability =  poisson.pmf(k, rate)


print(f"Probability(3 customers in an hour): {probability:.3f}")
```

Also answer: What would happen to the probability if the arrival rate increased to 8 customers per hour? Explain your answer

```python
# Changed the rate to 8
rate = 8
probability = poisson.pmf(k, rate)

print(f"Probability(3 customers in an hour with rate 8): {probability:.3f}")
```

if the increase changed from 5 to 8 customers per hour, the probability of exactly 3 customers arriving decreases due to the Poisson distribution becoming more spread out, making values farther from the mean less likely.

6. You are analyzing the distribution of weights of packages processed at a shipping center. The recorded sample weights (in kg) are:

10.5, 12.3, 11.7, 13.0, 10.8, 12.9, 11.5, 10.6, 12.1, 11.9

Write python code to:

Calculate the mean and standard deviation of the given data.

Plot a histogram of the given data.

Write the code to find the probability density function (PDF) of the normal distribution on the histogram using the calculated mean and standard deviation.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Given data
data = [10.5, 12.3, 11.7, 13.0, 10.8, 12.9, 11.5, 10.6, 12.1, 11.9]

# Calculating the mean
mean = np.mean(data)
std_dev = np.std(data)

# Standard deviation
print(f"Mean: {mean:.2f} kg")
print(f"Standard Deviation: {std_dev:.2f} kg")

# Plotting the histogram of the data
plt.figure(figsize=(8, 6))
plt.hist(data, bins=5, density=True, alpha=0.6, color='g', label='Histogram of Weights')

# Plotting the Probability Density Function
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean, std_dev)

plt.plot(x, p, 'k', linewidth=2, label=f'Normal PDF (mean={mean:.2f}, std={std_dev:.2f})')
plt.title('Histogram of Package Weights')
plt.xlabel('kg')
plt.ylabel('Density')
plt.legend()
plt.show()
```
