```python
import numpy as np
import pandas as pd
from scipy import stats

def calculate_mode(data):
    data = np.array(data)
    mode_result = stats.mode(data)
    return mode_result.mode

def analyze_central_tendency(data):
    data = np.array(data)
    results = {
        'Mean': np.mean(data),
        'Median': np.median(data),
        'Mode': calculate_mode(data),
    }
    return results

# This is a sample argument we can pass to the analyze_central_tendency function.
sample_data = [2, 3, 4, 4, 5, 5, 5, 6, 7, 8]
results = analyze_central_tendency(sample_data)

print("Central Tendency Analysis:")
for measure, value in results.items():
    print(f"{measure}: {value}")

```

Hint: What does Moderately Right-Skewed mean with respect to the data?

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

def analyze_distribution_shape(data):
    data = np.array(data)

    distribution_metrics = {
        'Skewness': stats.skew(data)
    }

    # Interpretation of skewness
    if distribution_metrics['Skewness'] > 1:
        skew_interpretation = "Highly Right-Skewed (Long right tail)"
    elif distribution_metrics['Skewness'] > 0.5:
        skew_interpretation = "Moderately Right-Skewed"
    elif distribution_metrics['Skewness'] > -0.5:
        skew_interpretation = "Approximately Symmetric"
    elif distribution_metrics['Skewness'] > -1:
        skew_interpretation = "Moderately Left-Skewed"



    # Visualization
    plt.figure(figsize=(12, 6))

    # Histogram with KDE
    plt.subplot(1, 2, 1)
    sns.histplot(data, kde=True)
    plt.title('Distribution Shape')
    plt.tight_layout()
    plt.show()


    distribution_metrics['Skewness'] = skew_interpretation
    return distribution_metrics


first_data = [1, 2, 2, 3, 3, 3, 4, 4, 5, 6, 7, 8, 10, 15, 20]
print(analyze_distribution_shape(first_data))

second_data = [1, 1, 2, 1, 1, 2, 5, 4, 4, 3, 3, 3, 2, 2, 1]
print(analyze_distribution_shape(second_data))

```

The first dataset, which is highly right-skewed, shows a long right tail due to a few larger values like 15 and 20 pulling the distribution to the right.

The second dataset exhibits moderately right-skewed distribution, with larger values having less significant impact than the first dataset.


Explain the significance of each dispersion metric in your own words

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def analyze_dispersion(data):
    data = np.array(data)
    dispersion_metrics = {
        'Variance': np.var(data),
        'Standard Deviation': np.std(data),
        'Interquartile Range (IQR)': np.percentile(data, 75) - np.percentile(data, 25)
    }


    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    sns.boxplot(x=data)
    plt.title('Box Plot - Dispersion Visualization')

    plt.subplot(1, 2, 2)
    sns.violinplot(x=data)
    plt.title('Violin Plot - Distribution Spread')

    plt.tight_layout()
    plt.show()

    return dispersion_metrics


sample_data = [10, 12, 15, 18, 20, 22, 25, 30, 35, 40]
results = analyze_dispersion(sample_data)
print("Dispersion Metrics:")
for metric, value in results.items():
    print(f"{metric}: {value:.2f}")
```

Variance indicates data point dispersion from the mean, with higher variance indicating more dispersion and lower variance indicating closerness. However, its squared units make interpretation challenging.

The standard deviation, the square root of the variance, provides a more interpretable measure of data spread, with small deviations indicating closeness to the mean and large deviations indicating greater spread.

The IQR is a reliable measure of data dispersion, spanning the 25th to 75th percentiles, ensuring it's not influenced by extreme values or outliers.
